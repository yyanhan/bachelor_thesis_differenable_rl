# show result in seaborn with data trained by different seeds

```python
def show(all, labels, n = 5000):
    """show result with seaborn

    Args:
        all (list of list): all the data
        labels (list of str): list of lables
        n (int, optional): num of steps. Defaults to 5000.
    """
    df = []
    for i in range(len(all)):
        reward = np.vstack(all[i])
        # print(reward)
        d = pd.DataFrame(reward).melt(var_name='episode',value_name='reward')
        # print("d",d)
        df.append(d)
        df[i]['method'] = labels[i]

    a=pd.concat(df, ignore_index = True)

    sns.lineplot(x="episode", y="reward", hue="method",style="method",data=a)
    plt.legend(loc="upper left")

    plt.xlabel("Episode")
    plt.ylabel("Reward")
    fn = "_".join(labels)
    plt.savefig(f"result-v-{n}-{fn}.png")
    plt.show()
```

# show relationships in matplot, with only one single seed

```python
def show_single(diff, g, method,seed):
    x = range(1, len(g)+1)
    plt.plot(x, diff, label = f"Method {method}, seed {seed}")
    plt.plot(x, g, label = f"PPO with Gym, seed {seed}")
    # plt.legend(loc="upper left")
    plt.legend(loc="lower right")
    plt.xlabel("Episode")
    plt.ylabel("Reward")
    plt.savefig(f"PPO-v-single-5000-{method}-{seed}.png")
    plt.show()
```
